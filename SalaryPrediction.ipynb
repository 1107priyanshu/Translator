```python
import pandas as pd
df=pd.read_csv('Salary Data.csv')
df
```

```python
df.isnull().sum()
```

```python
df[df.duplicated()]
```

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer(strategy='mean')
df['Age']=si.fit_transform(df[['Age']])
df['Years of Experience']=si.fit_transform(df[['Years of Experience']])
df['Salary']=si.fit_transform(df[['Salary']])
```

```python
si1=SimpleImputer(strategy='most_frequent')
df['Gender']=si1.fit_transform(df[['Gender']]).ravel()
df['Education Level']=si1.fit_transform(df[['Education Level']]).ravel()
df['Job Title']=si1.fit_transform(df[['Job Title']]).ravel()
```

```python
df.isnull().sum()
```

```python
df = df.drop_duplicates(keep='first')
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe=OrdinalEncoder()
df['Education Level']=oe.fit_transform(df[['Education Level']])
df['Gender']=oe.fit_transform(df[['Gender']])
df['Job Title']=oe.fit_transform(df[['Job Title']])
```

```python
x=df.iloc[:,0:-1]
y=df.iloc[:,-1]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
```

```python
from sklearn.metrics import r2_score
r2_score(y_test,pred)
```

```python
from yellowbrick.model_selection import ValidationCurve
from sklearn.ensemble import RandomForestRegressor
import matplotlib.pyplot as plt
rfc=RandomForestRegressor()
rfc.fit(x_train,y_train)
pred=rfc.predict(x_test)
r2_score(pred,y_test)
```

```python
num_est=[100,200,400,500,700,850,900,100]
val=ValidationCurve(RandomForestRegressor(),x=x_train,y=y_train,param_name='n_estimators',param_range=num_est,scoring='r2',cv=5)
val.fit(x_train,y_train)
val.show()
```

```python
depth=[11,8,15,12,20,22,25]
val=ValidationCurve(RandomForestRegressor(),x=x_train,y=y_train,param_name='max_depth',param_range=depth,scoring='r2',cv=5)
val.fit(x_train,y_train)
val.show()
```

```python
min=[3,6,4,5,8,12,15,17]
val=ValidationCurve(RandomForestRegressor(),x=x_train,y=y_train,param_name='min_samples_split',param_range=min,scoring='r2',cv=5)
val.fit(x_train,y_train)
val.show()
```

```python
rfr=RandomForestRegressor(n_estimators=700,max_depth=20,min_samples_split=6)
rfr.fit(x_train,y_train)
pred1=rfr.predict(x_test)
r2_score(pred1,y_test)
```

```python
import joblib
joblib.dump(rfr,'model.pkl')
joblib.dump(oe,'encoder.pkl')
```

```python
def randomforestmodel(age,gender,edu,job,exp):
  rfr=joblib.load('model.pkl')
  oe=joblib.load('encoder.pkl')
  input=pd.DataFrame([[age,gender,edu,job,exp]])
  pred2=rfr.predict(input)[0]
  print("Predicted Salary=",pred2)
age=int(input("Enter age:"))
gender=int(input("Enter gender(standerize):"))
edu=int(input("Enter edu(standerize):"))
job=int(input("Enter job(standerize):"))
exp=int(input("Enter exp:"))
randomforestmodel(age,gender,edu,job,exp)
```
